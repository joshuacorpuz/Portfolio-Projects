```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataOG = pd.read_csv("India_budget_2021.csv")
data = dataOG
data.head()
```

```python
#Check all of the data first
print(data)
```

```python
#Drop all NaN values
data.dropna()
```

```python
#Create a new dataframe with the main departments' data and then add the combined funds of all the non-main departments as 'others'
data = data.loc[[0,8,11,14,18,23,41,42,43],:]
row = pd.DataFrame({'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}, index=[0])
data = pd.concat([data, row], ignore_index = True)
print(data)
```

```python
#Use a bar graph to illustrate how much funds each department receives
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

```python
#Use a pie graph to illustrate the percentage of the overall budget that each department is receiving. Here we can see that the Ministry of Finance is receiving a significant
# chunk of the country's budget
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=30)
plt.show()
```
