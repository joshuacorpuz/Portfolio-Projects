```python
import pandas as pd
import os
```

```python
dataset = pd.read_excel( "D:\\Download\\A2J Survey to SRLs All Responses.xlsx", sheet_name="Survey SRLs All Response Edited")
dataset
```

```python
dataset_edit = dataset.copy()
dataset_edit
```

```python
list(dataset_edit.columns)
```

```python
id_vars = list(dataset_edit.columns)[0]
value_vars = list(dataset_edit.columns)[1 : ]
#value_vars
```

```python
dataset_melted = dataset_edit.melt(id_vars=id_vars, value_vars=value_vars, var_name="Question + Sub-Question", value_name="Answer")
dataset_melted
```

```python
questions_import = pd.read_excel( "D:\\Download\\A2J Survey to SRLs All Responses.xlsx", sheet_name="Questions")
questions_import
```

```python
questions = questions_import.copy()
questions
```

```python
dataset_merged = pd.merge(left=dataset_melted ,right=questions ,how="left" ,left_on="Question + Sub-Question" ,right_on="Question + Subquestion")
print("Original Data", len(dataset_melted))
print("Merged Data", len(dataset_merged))
dataset_merged
```

```python
respondents = dataset_merged.groupby("Question + Subquestion")["Person ID"].count().reset_index()
respondents
```

```python
respondents = dataset_merged[dataset_merged["Answer"].notna()]
respondents = respondents.groupby("Question + Subquestion")["Person ID"].nunique().reset_index()
respondents.rename(columns={"Person ID":"Respondents"}, inplace=True)
#respondents
print(respondents.to_string())
```

```python
dataset_merged_two = pd.merge(left=dataset_melted ,right=respondents ,how="left" ,left_on="Question + Sub-Question" ,right_on="Question + Subquestion")
print("Original", len(dataset_merged))
print("New", len(dataset_merged_two))
print(dataset_merged_two[999:1000].to_string())
```
