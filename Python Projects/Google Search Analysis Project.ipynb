```python
# import libraries

import numpy as np
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
# Looking at the interest in the keyword and showing the top 10 regions

trends.build_payload(kw_list=['Demon Slayer'])
data = trends.interest_by_region()
data = data.sort_values(by = 'Demon Slayer', ascending=False)
data = data.head(10)
print(data)
```

```python
# Creating a bar graph to illustrate the difference in interest between the regions
# This shows that the Philippines and Mongolia has the highest interest in searching up Demon Slayer

data.reset_index().plot(x='geoName', y='Demon Slayer', figsize=(12,8), kind='bar', legend = None)
plt.style.use('dark_background')
plt.xlabel("Country")
plt.ylabel("Interest Level")
plt.title("Region Interest in Searching Demon Slayer in Google")
plt.show()
```

```python
# Creating a line graph showing difference in the amount of google searches over the years
# The first spike in late 2019 can be attributed to the airing of the first season of the anime, Demon Slayer
# The second gigantic spike is attributed to the western release of the Demon Slayer movie that was heavily anticipated at the time 
# The final gigantic spike is attributed to the airing of the second season of the anime over a span of 11 weeks

data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Demon Slayer'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(12,8))
data['Demon Slayer'].plot()
plt.style.use('dark_background')
plt.title('Total Google Searches for Demon Slayer')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
